```python
from pymysql import Connection
import matplotlib.pyplot as plt
conn = Connection(
    host="localhost",
    port=3306,
    user="root",
    password="123"
)
cursorH = conn.cursor()
cursorN0 = conn.cursor()
cursorN1 = conn.cursor()
cursorN2 = conn.cursor()
conn.select_db("citydata")
```

```python
cursorH.execute("select X,Y from atlantaoutput where Hotel = 'Y';")
H = cursorH.fetchall()
cursorN0.execute("select X,Y,cluster from atlantaoutput  where Hotel = 'N' and cluster = 0;")
N0 = cursorN0.fetchall()
cursorN1.execute("select X,Y,cluster from atlantaoutput  where Hotel = 'N' and cluster = 1;")
N1 = cursorN1.fetchall()
cursorN2.execute("select X,Y,cluster from atlantaoutput  where Hotel = 'N' and cluster = 2;")
N2 = cursorN2.fetchall()

Xh = []
Yh = []
Xn0 = []
Yn0 = []
Xn1 = []
Yn1 = []
Xn2 = []
Yn2 = []

for i in H:
    Xh.append(i[0])
    Yh.append(i[1])
for i in N0:
    Xn0.append(i[0])
    Yn0.append(i[1])
for i in N1:
    Xn1.append(i[0])
    Yn1.append(i[1])
for i in N2:
    Xn2.append(i[0])
    Yn2.append(i[1])


plt.style.use('fivethirtyeight')

plt.figure(figsize=(10,10))
plt.scatter(Xh, Yh, s=100, c="red",label="Hotel",marker='*')
plt.scatter(Xn0, Yn0, s=100, c="blue",label="Attraction in cluster 0")
plt.scatter(Xn1, Yn1, s=100, c="green",label="Attraction in cluster 1")
plt.scatter(Xn2, Yn2, s=100, c="purple",label="Attraction in cluster 2")
plt.xlabel("Longitude",fontsize = 20)
plt.ylabel("latitude",fontsize = 20)
plt.title("Clusters in Atlanta", fontsize = 20)
plt.legend()
plt.show
```
